```python
import numpy as np

#  NumPy es una librerÃ­a de Python especializada en el cÃ¡lculo numÃ©rico y el anÃ¡lisis de datos, especialmente para un gran volumen de datos.

#
## la siguiente funciÃ³n divide el test de datos en un conjunto de prueba y otro de entrenamiento
# @data: el set de datos
# @test_ratio: porciento del conjunto de instancias aleatorias seleccionadas (por lo general el 20% de los datos 
# o menos si el conjunto es muy grande)
# 

def split_train_test(data,test_ratio):
    shuffled_indices = np.random.permutation(len(data))# realiza una nueva distribuciÃ³n aleatoria de la data. @len(data) es longitud de la data
    print(shuffled_indices)
    test_set_size = int(len(data) * test_ratio)# producto de la longitud de la data por la fracciÃ³n de la data (generalmente 20%)
    test_indices = shuffled_indices[:test_set_size] # Los datos organizados aleatoriamente se dividen (slice 20%) desde 0 hasta el valor de test_set_size 
    train_indices = shuffled_indices[test_set_size:] #Los datos organizados aleatoriamente se dividen (slice 80%) desde test_set_size hasta el valor final
    return data.iloc[train_indices], data.iloc[test_indices] # .iloc es un mÃ©todo de Pandas utilizado para la indexaciÃ³n basada en ubicaciones de nÃºmeros enteros. Le permite seleccionar filas y columnas por sus posiciones enteras.
```

```python
#En este caso se toma el 20% (0.2)del conjunto de datos. Si el conjunto de entrenamiento es muy grande, se toma menos
# Generando el conjunto de prueba y entrenamiento
train_set, test_set = split_train_test(housing,0.2)
```

```python
#Posible implementaciÃ³n para lograr un conjunto de prueba estable
from zlib import crc32

def test_set_check(identifier,test_ratio):
    return crc32(np.int64(identifier)) & 0xffffffff < test_ratio * 2**3

def split_train_test_by_id(data,test_ratio,id_column):
    ids = data[id_column]
    in_test_set = ids.apply(lambda id_: test_set_check(id_,test_ratio))
    return data.loc[~in_test_set],data.loc[in_test_set]
```

```python
#Variante 1: Dado que el set de datos no tiene un identificador, se crea un Ã­ndice utilizando el nÃºmero de cada fila
#Utilizar el nÃºmero de fila implica que los datos nuevos se aÃ±adan al final de la fila
housing_with_id=housing
housing_with_id = housing.reset_index()#aÃ±ade una columna index
train_set, test_set = split_train_test_by_id(housing_with_id, 0.2, "index")
```

```python
#Variante 2: Utilizar las columnas o campos mÃ¡s estables de la data para crear un identificador Ãºnico
#En este caso particular, se puede garantizar que la latitud y la longitud se mantendrÃ¡n estables por siglos, 
#por eso pueden utilizarse par construir un Ã­ndice Ãºnico
#housing_with_id = housing.reset_index()#aÃ±ade una columna index
housing_with_id=housing
housing_with_id["id"] = housing["longitude"] * 1000 + housing["latitude"]
train_set, test_set = split_train_test_by_id(housing_with_id,0.2,"id")
```
