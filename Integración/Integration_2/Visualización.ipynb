```python
# %pip install matplotlib
# Recordar y prácticar
# La función zip() combina dos listas en una lista de tuplas
# La función dict() convierte una lista de tuplas en un diccionario
# El método random.uniform(a,b) devuelve un número float decimal aleatorio entre a y b, excluyendo ambos extremos
# Generador de listas por comprensión: [random.uniform(1, 10) for _ in range(10)]
# random.uniform(1, 10) : Genera un número aleatorio entre 1 y 10
# for _ in range(10) : Es un bucle que repite 10 veces. Usas la variable anónima _ cuando no necesitas usar el valor de un contador (es como
# si dijeras al programa: Olvidate de todo, solo repite 10 veces)
# Los corchetes [...] indican que todo el bucle se ejecute en una lista

def create_list():
    # Generar una lista de números aleatorios
    lista_numeros = [random.uniform(1, 10) for _ in range(10)]
    return lista_numeros

def calcular_mediana(lista):
    return statistics.median(lista)

def calcular_stdev(lista):
    return statistics.stdev(lista)

def main():
    lista = create_list()
    # Calcular y mostrar la mediana y desviación típica
    print("Lista de números:", lista)
    print("Mediana:", calcular_mediana(lista))
    print("Desviación Típica:", calcular_stdev(lista))

    # Histograma
    # La función hist() de matplotlib crea un histograma a partir de los datos de la lista
    # figsize especifica el tamaño de la figura
    # bins especifica el número de intervalos en el histograma
    # La función grid() añade una cuadrícula al gráfico
    # La función show() muestra el gráfico
    # La función title() establece el título del gráfico
    plt.figure(figsize=(10, 5))
    plt.hist(lista, bins=10, edgecolor='black')
    plt.title("Histograma de Números")
    plt.xlabel("Valor")
    plt.ylabel("Frecuencia")
    plt.grid(True)
    plt.show()

    # Histogram with mode and median lines
    # La función hist() de matplotlib crea un histograma a partir de los datos de la lista
    # edgecolor='black' añade un borde 
    # alpha=0.7 establece la transparencia del histograma
    # La función axvline() dibuja una línea vertical en la posición de la mediana
    # La función legend() añade una leyenda al gráfico
    # La función grid() añade una cuadrícula al gráfico
    plt.figure(figsize=(10, 5))
    plt.hist(lista, bins=10, edgecolor='black', alpha=0.7)
    plt.axvline(calcular_mediana(lista), color='red', linestyle='-', linewidth=2, label=f'Mediana: {calcular_mediana(lista):.2f}')
    plt.title("Histograma con Mediana")
    plt.xlabel("Valor")
    plt.ylabel("Frecuencia")
    plt.legend()
    plt.grid(True)
    plt.show()

    # Boxplot es una representación gráfica de la distribución de un conjunto de datos
    # que muestra la mediana, los cuartiles y los valores atípicos
    # La función boxplot() de matplotlib crea un boxplot a partir de los datos de la lista
    # La función title() establece el título del gráfico
    # La función ylabel() establece la etiqueta del eje y
    # La función grid() añade una cuadrícula al gráfico
    # La función show() muestra el gráfico
    # La función annotate() añade una anotación al gráfico   

    # Boxplot
    plt.figure(figsize=(6, 4))
    plt.boxplot(lista)
    plt.title("Diagrama de Caja (Boxplot)")
    plt.ylabel("Valor")
    plt.grid(True)
    plt.annotate(f'Mediana: {calcular_mediana(lista):.2f}', xy=(1, calcular_mediana(lista)), xytext=(1.1, calcular_mediana(lista) + 0.5),
                 arrowprops=dict(facecolor='black', arrowstyle='->'), fontsize=10, color='red')
    plt.show()


    # Scatter plot
    plt.figure(figsize=(10, 5))
    plt.plot(lista, marker='o', linestyle='-', alpha=0.7)
    plt.title("Secuencia de Números Aleatorios")
    plt.xlabel("Índice")
    plt.ylabel("Valor")
    plt.grid(True)
    plt.show()

if __name__ == "__main__":
    import random
    import statistics
    import matplotlib.pyplot as plt
    main()

```
