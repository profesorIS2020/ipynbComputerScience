```python
# %pip install matplotlib
# El m칠todo random.uniform(a,b) devuelve un n칰mero float decimal aleatorio entre a y b, excluyendo ambos extremos
# Generador de listas por comprensi칩n: [random.uniform(1, 10) for _ in range(10)]
# random.uniform(1, 10) : Genera un n칰mero aleatorio entre 1 y 10
# for _ in range(10) : Es un bucle que repite 10 veces. Usas la variable an칩nima _ cuando no necesitas usar el valor de un contador (es como
# si dijeras al programa: Olvidate de todo, solo repite 10 veces)
# Los corchetes [...] indican que todo el bucle se ejecute en una lista

def create_list():
    # Generar una lista de n칰meros aleatorios
    lista_numeros = [random.uniform(1, 10) for _ in range(10)]
    return lista_numeros

def calcular_mediana(lista):
    return statistics.median(lista)

def calcular_stdev(lista):
    return statistics.stdev(lista)

def main():
    lista = create_list()
    # Calcular y mostrar la mediana y desviaci칩n t칤pica
    print("Lista de n칰meros:", lista)
    print("Mediana:", calcular_mediana(lista))
    print("Desviaci칩n T칤pica:", calcular_stdev(lista))

    # Histograma
    # La funci칩n hist() de matplotlib crea un histograma a partir de los datos de la lista
    # figsize especifica el tama침o de la figura
    # bins especifica el n칰mero de intervalos en el histograma
    # La funci칩n grid() a침ade una cuadr칤cula al gr치fico
    # La funci칩n show() muestra el gr치fico
    # La funci칩n title() establece el t칤tulo del gr치fico
    plt.figure(figsize=(10, 5))
    plt.hist(lista, bins=10, edgecolor='black')
    plt.title("Histograma de N칰meros")
    plt.xlabel("Valor")
    plt.ylabel("Frecuencia")
    plt.grid(True)
    plt.show()

    # Histogram with mode and median lines
    # La funci칩n hist() de matplotlib crea un histograma a partir de los datos de la lista
    # edgecolor='black' a침ade un borde 
    # alpha=0.7 establece la transparencia del histograma
    # La funci칩n axvline() dibuja una l칤nea vertical en la posici칩n de la mediana
    # La funci칩n legend() a침ade una leyenda al gr치fico
    # La funci칩n grid() a침ade una cuadr칤cula al gr치fico
    plt.figure(figsize=(10, 5))
    plt.hist(lista, bins=10, edgecolor='black', alpha=0.7)
    plt.axvline(calcular_mediana(lista), color='red', linestyle='-', linewidth=2, label=f'Mediana: {calcular_mediana(lista):.2f}')
    plt.title("Histograma con Mediana")
    plt.xlabel("Valor")
    plt.ylabel("Frecuencia")
    plt.legend()
    plt.grid(True)
    plt.show()

    # Boxplot es una representaci칩n gr치fica de la distribuci칩n de un conjunto de datos
    # que muestra la mediana, los cuartiles y los valores at칤picos
    # La funci칩n boxplot() de matplotlib crea un boxplot a partir de los datos de la lista
    # La funci칩n title() establece el t칤tulo del gr치fico
    # La funci칩n ylabel() establece la etiqueta del eje y
    # La funci칩n grid() a침ade una cuadr칤cula al gr치fico
    # La funci칩n show() muestra el gr치fico
    # La funci칩n annotate() a침ade una anotaci칩n al gr치fico   

    # Boxplot
    plt.figure(figsize=(6, 4))
    plt.boxplot(lista)
    plt.title("Diagrama de Caja (Boxplot)")
    plt.ylabel("Valor")
    plt.grid(True)
    plt.annotate(f'Mediana: {calcular_mediana(lista):.2f}', xy=(1, calcular_mediana(lista)), xytext=(1.1, calcular_mediana(lista) + 0.5),
                 arrowprops=dict(facecolor='black', arrowstyle='->'), fontsize=10, color='red')
    plt.show()


    # Scatter plot
    plt.figure(figsize=(10, 5))
    plt.plot(lista, marker='o', linestyle='-', alpha=0.7)
    plt.title("Secuencia de N칰meros Aleatorios")
    plt.xlabel("칈ndice")
    plt.ylabel("Valor")
    plt.grid(True)
    plt.show()

if __name__ == "__main__":
    import random
    import statistics
    import matplotlib.pyplot as plt
    main()

```

```python
#%pip install matplotlib

# Diccionario por comprensi칩n

def get_names():
    # Generar un diccionario de nombres y edades
    nombres = [
        "Nicol치s", "Carlos", "Luis Gabriel", "Alfonso", "Jos칠 Ignacio", 
        "Emilio Eugenio", "Diana del Carmen", "Pablo", "Daniel", 
        "Maria del Sol", "Jessica Ximena del Pilar", "Rodrigo", "Jorge", 
        "Diana Fernanda", "Luis Fernando", "Javier", "Pedro", "Paloma", 
        "Marcos", "Enrique", "Guiomar", "Daira", "Santiago", "Luis Enrique"
    ]
    return nombres

def get_grades():
    nombres = get_names()
    grades = {name: round(random.uniform(1, 10),2) for name in nombres}
    return grades

def calcular_mediana(lista):
    return statistics.median(lista)

def calcular_stdev(lista):
    return statistics.stdev(lista)

def create_hist(lista):
    # Histograma
    # La funci칩n hist() de matplotlib crea un histograma a partir de los datos de la lista
    # figsize especifica el tama침o de la figura
    # bins especifica el n칰mero de intervalos en el histograma
    # La funci칩n grid() a침ade una cuadr칤cula al gr치fico
    # La funci칩n show() muestra el gr치fico
    # La funci칩n title() establece el t칤tulo del gr치fico
    plt.figure(figsize=(10, 5))
    plt.hist(lista, bins=10, edgecolor='black')
    plt.title("Histograma de N칰meros")
    plt.xlabel("Valor")
    plt.ylabel("Frecuencia")
    plt.grid(True)
    plt.show()

def create_hist_with_median(lista):
    # Histogram with mode and median lines
    # La funci칩n hist() de matplotlib crea un histograma a partir de los datos de la lista
    # edgecolor='black' a침ade un borde 
    # alpha=0.7 establece la transparencia del histograma
    # La funci칩n axvline() dibuja una l칤nea vertical en la posici칩n de la mediana
    # La funci칩n legend() a침ade una leyenda al gr치fico
    # La funci칩n grid() a침ade una cuadr칤cula al gr치fico
    plt.figure(figsize=(10, 5))
    plt.hist(lista, bins=10, edgecolor='black', alpha=0.7)
    plt.axvline(calcular_mediana(lista), color='red', linestyle='-', linewidth=2, label=f'Mediana: {calcular_mediana(lista):.2f}')
    plt.title("Histograma con Mediana")
    plt.xlabel("Valor")
    plt.ylabel("Frecuencia")
    plt.legend()
    plt.grid(True)
    plt.show()

def create_boxplot(lista):  
    # Boxplot es una representaci칩n gr치fica de la distribuci칩n de un conjunto de datos
    # que muestra la mediana, los cuartiles y los valores at칤picos
    # La funci칩n boxplot() de matplotlib crea un boxplot a partir de los datos de la lista
    # La funci칩n title() establece el t칤tulo del gr치fico
    # La funci칩n ylabel() establece la etiqueta del eje y
    # La funci칩n grid() a침ade una cuadr칤cula al gr치fico
    # La funci칩n show() muestra el gr치fico
    # La funci칩n annotate() a침ade una anotaci칩n al gr치fico   

    # Boxplot
    plt.figure(figsize=(6, 4))
    plt.boxplot(lista)
    plt.title("Diagrama de Caja (Boxplot)")
    plt.ylabel("Valor")
    plt.grid(True)
    plt.annotate(f'Mediana: {calcular_mediana(lista):.2f}', xy=(1, calcular_mediana(lista)), xytext=(1.1, calcular_mediana(lista) + 0.5),
                 arrowprops=dict(facecolor='black', arrowstyle='->'), fontsize=10, color='red')
    plt.show()

def create_scatter_plot(lista):
    # Scatter plot
    plt.figure(figsize=(10, 5))
    plt.plot(lista, marker='o', linestyle='-', alpha=0.7)
    plt.title("Secuencia de N칰meros Aleatorios")
    plt.xlabel("칈ndice")
    plt.ylabel("Valor")
    plt.grid(True)
    plt.show()


def main():
    # Generar un diccionario de nombres y edades
    grades = get_grades()
    print("Diccionario de Nombres y Notas:")
    print(f"{'Nombre':<29} {'Nota':>0}")
    print("-" * 20)
    # Mostrar el diccionario de nombres y notas
    for name, grade in grades.items():
        print(f"{name:<10} {grade:>0}")
    
    print("-" * 20)
    # Calcular la media
    media = sum(grades.values()) / len(grades)
    print("Media:", media)
    print("-" * 20)
    # Calcular la mediana
    mediana = statistics.median(grades.values())
    print("Mediana:", mediana)
    print("-" * 20)
    # Calcular la desviaci칩n t칤pica
    stdev = statistics.stdev(grades.values())
    print("Desviaci칩n T칤pica:", stdev)

    print("-" * 20)
    # Crear un histograma de las notas
    create_hist(grades.values())
    print("-" * 20)

    # Crear un histograma con la mediana   
    create_hist_with_median(grades.values())
    print("-" * 20) 
    # Crear un boxplot
    create_boxplot(grades.values())    
    print("-" * 20)
    # Crear un scatter plot
    create_scatter_plot(grades.values())
    print("-" * 20)

if __name__ == "__main__":
    import random
    import statistics
    main()
```
